```python
from sklearn.linear_model import LinearRegression
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error


df = pd.read_csv("csv/clean_dataset.csv")
```

```python
df.head()
```

```python
X_df = df.drop(["today_sold_units", "date"], axis=1)
```

```python
X_df.head()
```

```python
y_df = df["today_sold_units"]
```

```python
y_df.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X_df, y_df, test_size=0.25
)
```

```python
print(X_train.value_counts().sum())
```

```python
y_test.value_counts().sum()
```

```python
model = LinearRegression()
```

```python
model.fit(X_train, y_train)
```

```python
X_test.head()
```

```python
y_prediction = model.predict(X_test)  # [[103.312634, 0, 114, 4, 1, 1.287112]]
print(y_prediction)
```

```python
import numpy as np
mae = mean_absolute_error(y_test, y_prediction)
rmse = np.sqrt(mean_squared_error(y_test, y_prediction))

print("MAE:", mae)
print("RMSE:", rmse)
```
